In this Jupyter notebook, we will be building a classification model for the breast cancer toy data set in scikit learn using the random forest algorithm.

 ## 1. Import libraries

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
```

## 2. Load the *breast cancer* data set

```python
bc = datasets.load_breast_cancer()
bc
```

## 3. Input features
The ***breast cancer*** data set contains 30 input features and 1 output variable (the class label).

### 3.1. Input features

```python
print(bc.feature_names)
```

### 3.2. Output features

```python
print(bc.target_names)
```

```python
print(bc.target)
```

## 4. About the data

### 4.1. Input features

```python
bc.data
```

### 4.2. Output variable or the Class label

```python
bc.target
```

### 4.3. Assigning *input* and *output* variables
Let's assign the 30 input variables to X and the output variable (class label) to Y

```python
X = bc.data
Y = bc.target
```

### 4.4. Let's examine the data dimension

```python
X.shape
```

```python
Y.shape
```

## 5. Build Classification Model using Random Forest

```python
clf = RandomForestClassifier()
```

```python
clf.fit(X, Y)
```

## 6. Feature Importance

```python
print(clf.feature_importances_)
```

## 8. Data split (80/20 ratio)

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
X_train.shape, Y_train.shape
```

```python
X_test.shape, Y_test.shape
```

## 9. Rebuild the Random Forest Model

```python
clf.fit(X_train, Y_train)
```

### 9.1. Performs prediction on single sample from the data set

```python
print(clf.predict(X[[0]]))
```

```python
print(clf.predict_proba(X[[0]]))
```

### 9.2. Performs prediction on the test set

#### Predicted class labels

```python
print(clf.predict(X_test))
```

#### Actual class labels

```python
print(Y_test)
```

## 10. Model Performance

```python
print(clf.score(X_test, Y_test))
```
